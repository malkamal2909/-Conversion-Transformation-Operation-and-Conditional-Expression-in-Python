# **Definition**

**Expression** is a sentence with a minimum of two numbers and at least one math operation.

**Operator** is special symbols that designate that some sort of computation should be performed.

**Operand** is the values that an operator acts.

**Example: 2 + 3** 

- sentence 2 + 3 is an expression

- symbol (+) is an operator

- number 2 and 3 are operands

2. Exponents

# **Assignment Operator**

## Operator (+=)


```python
a = 3
a += 6 # it is equivalent to a = a+6
a
```

## Operator (-=)

```python
b = 8
b -= 7 # it is equivalent to b = b-7
b
```

## Operator (*=)

```python
c = 2
c *= 4 # it is equivalent to c = c*4
c
```

## Operator (/=)

```python
d = 6
d /= 3 # it is equivalent to d = d/3
d
```

## Operator (%=)

```python
e = 4
e %= 3 # it is equivalent to e = e%3
e
```

## Operator (//=)

```python
f = 6
f //= 4 # it is equivalent to f = f//4
f
```
