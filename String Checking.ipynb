Cheking on String

In the string checking category we will check the boolean of a string

isupper()

The isupper() method returns True if all the characters are in upper case, otherwise False.

Numbers, symbols and spaces are not checked, only alphabet characters

```python
a = "DATA SCIENCE"
```

```python
a.isupper()
```

```python
b = 'data science'
```

```python
b.isupper()
```

islower()

The islower() method returns True if all the characters are in lower case, otherwise False.

Numbers, symbols and spaces are not checked, only alphabet characters.



```python
a = "data science"
```

```python
a.islower()
```

```python
b = "Data Science"
```

```python
b.islower()
```

We can perform operations on the result of the operation (method chain)

```python
print("Science".upper().lower())
```

```python
print("Science".lower().upper())
```

```python
print("Science".upper().lower().islower())
```

```python
print("Science".upper().lower().isupper())
```

isalpha()

The isalpha() method returns True if all the characters are alphabet letters (a-z).

Example of characters that are not alphabet letters: (space)!#%&? etc.

```python
print("Data".isalpha())
```

```python
print("Ibu2".isalpha())
```

Because there is a number, then the output returns False

isalnum()

The isalnum() method returns True if all characters in the string are alphanumeric (either alphabets or numbers). If not, it returns False.

```python
print('Tahun2022'.isalnum())
```

```python
print('Data 2022'.isalnum())
```

```python
print('2022'.isalnum())
```

```python
print('Data'.isalnum())
```

isdecimal()

The isdecimal() method returns True if all the characters are decimals (0-9).

This method is used on unicode objects.

```python
print("Tahun2022".isdecimal())
```

```python
print("2022".isdecimal())
```

isspace()

The isspace() method returns True if all the characters in a string are whitespaces, otherwise False.

```python
print('\n'.isspace())
```

```python
print(' '.isspace())
```

istitle()

The istitle() method returns True if all words in a text start with a upper case letter, AND the rest of the word are lower case letters, otherwise False.

Symbols and numbers are ignored.



```python
print("Data Science".istitle())
```

```python
print("Malka Rusyd".istitle())
```

```python
print("maulana ishaq".istitle())
```
