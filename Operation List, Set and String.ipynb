# **len**
Returns the length of an object (list, set or string)

```python
a = 'Data Science'
len(a)
```

```python
a = (2,7,1,1,2,4,8,'eleven')
len(a)
```

```python
# return the lenght string in the tuple
len(a[-1])
```

```python
a = ['eight','eleven','ten']
len(a)
```

```python
# the data in the set is unique so the duplicate data will only appear once
a = {2,7,1,1,2,4,8,'eleven'}
len(a)
```

# **count**
Returns the number of elements with the specified value

```python
a = "Data Science"
a.count('a')
```

```python
a = (2,7,1,1,2,4,8,'eleven')
a.count('eleven')
```

```python
a = [2,7,1,1,2,4,8,'eleven','ten']
a.count(2)
```

```python
# return the number of element in the string in the list
a[-2].count('e')
```

```python
# object 'set' does not have attribute 'count' because 'set' is uniqe so each data must be one
a = {2,7,1,1,2,4,8,'eleven'}
a.count(2)
```

# **reverse**
reverses the sorting order of the elements of list.

reverse cannot be used to string, tuples or set.

```python
a = [2,7,1,1,2,4,8,11]
a.reverse()
print(a)
```

```python
a = [3,'one',9,2,'ten']
a.reverse()
print(a)
```

```python
a = ['Data Science','Data Analyst','Python']
a.reverse()
print(a)
```

# **sort**
Sorts the list ascending by default. Tuple, set, and string cannot support sort.

String in list sorted alphabetically from the first character. if the first character is the same then sorted by the second character, and so on.

```python
a = [2,7,1,1,2,4,8,11]
a.sort()
print(a)
```

```python
a = ['Data Science','Data Analyst','Python']
a.reverse()
print(a)
```

```python
# comparison not supported between instances of string and number (integer or float)
a = [3,'one',9,2,'ten']
a.sort()
print(a)
```

```python
# sort descending
a = [2,7,11,2,3,4,8,11,9,9]
a.sort(reverse=True)
print(a)
```

```python
# case sensitive
a = ['data','Python','coding','string','List']
a.sort()
print(a)
```

```python
# sort descending by ignoring case sensitive without changing the contents of the list
a = ['data','Python','coding','string','List']
a.sort(key=str.lower,reverse=True)
print(a)
```

```python
# the numbers type string in list are sorted according to the sorting rules on the string 
a = ['1','16','55','151','213','3']
a.sort()
print(a)
```

# **max**
Returns the item with the highest value, or the item with the highest value in an iterable.

If the values are strings, an alphabetically comparison is done.

```python
a = "Data Science"
max(a)
```

```python
a = (2,7,1,1,2,4,8,11)
max(a)
```

```python
a = {2,7,1,1,2,4,8}
max(a)
```

```python
# max string in a list. return the name with the highest value, ordered alphabetically. same as in sorting.
a = ['eight','eleven','ten']
max(a)
```

```python
# same as sort. comparison not supported between instances of string and number (integer or float)
a = [2,7,1,1,2,4,8,'eleven']
max(a)
```

# **min**
Returns the item with the lowest value, or the item with the lowest value in an iterable.

If the values are strings, an alphabetically comparison is done.

```python
a = "python"
min(a)
```

```python
# Case sensitive. The alphabetical order will be counted starting with the capital letters first
a = 'Python'
min(a)
```

```python
a = [2,7,1,1,2,4,8,11]
min(a)
```

```python
a = {2,7,1,1,2,4,8}
min(a)
```

```python
# return min of element in the string in the tuple
a = (2,7,1,1,2,4,8,'eleven')
min(a[-1])
```

```python
# same as sort. comparison not supported between instances of string and number (integer or float)
a = [2,7,1,1,2,4,8,'eleven']
min(a)
```

# **in**
Returns True if a sequence with the specified value is present in the object

```python
a = [1,2,3,4,5]
1 in a
```

```python
a = 'Data Science'
'c' in a
```

```python
# case sensitive
a = 'Data Science'
'd' in a
```

# **not in**
Returns True if a sequence with the specified value is not present in the object

```python
a = (1,2,3,4,5)
1 not in a
```

```python
a = 'Data Science'
'd' not in a
```

# **range**
The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

syntax : range(start, stop, step)

```python
for i in range(3):
  print(i)
```

```python
for i in range (2,6):
  print(i)
```

```python
for i in range (2,10,2):
  print(i)
```
